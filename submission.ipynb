```python
import pandas as pd
```

```python
df = pd.read_csv('InceptionResNetV2_predictions_2.csv', index_col = 0)
df
```

```python
prediction = []
for row in df.itertuples():
    cur = []
    for i, val in enumerate(row[1:]):
        if val > 0.05:
            cur.append(i)
    prediction.append(' '.join([str(num) for num in cur]))
```

```python
prediction
```

```python
df_res = pd.DataFrame(columns = ['Id', 'Predicted'])
```

```python
df_res['Predicted'] = prediction
```

```python
df2['Predicted'] = prediction
df2.head()
```
